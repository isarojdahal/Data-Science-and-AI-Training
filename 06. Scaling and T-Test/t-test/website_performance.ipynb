```python
from scipy import stats
import numpy as np

# Group A: Page load time (seconds) with Old Server
group_a = [3.2, 3.5, 2.9, 3.1, 3.4] 
# Group B: Page load time (seconds) with New Server
group_b = [2.1, 2.3, 2.0, 1.9, 2.2]

t_stat, p_val = stats.ttest_ind(group_a, group_b)

print(f"P-Value: {p_val:.4f}")

if p_val < 0.05:
    print("The new server significantly improved speed!")
else:
    print("The speed difference is not statistically significant.")
```
