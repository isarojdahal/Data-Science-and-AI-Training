```python
from transformers import pipeline
import scipy

# Load the Bark model
pipe = pipeline("text-to-speech", model="suno/bark-small")

# Generate (this might take a moment to download the first time)
output = pipe("Hello! [laughter] This is a much more natural voice.")

# Save the output
scipy.io.wavfile.write("bark_out.wav", rate=output["sampling_rate"], data=output["audio"])
```
