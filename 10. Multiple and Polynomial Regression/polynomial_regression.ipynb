```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python

df = pd.DataFrame({
    'Temperature':[1, 2, 3, 4, 5, 6, 7, 8],
    'Energy Consumption':[2, 5, 9, 15, 14, 10, 5, 1]
})
```

```python
sns.scatterplot(data=df, x="Temperature",y="Energy Consumption")
```

```python
x = df[['Temperature']]
y = df['Energy Consumption']
x_poly = PolynomialFeatures(degree=2).fit_transform(x)
print(x_poly)
```

```python
model = LinearRegression().fit(x_poly,y)
print(model.intercept_) 
print(model.coef_)


# for value 2. 
# energy_consumption_predicted = model.predict([[1,2,4]])

# more better way.
energy_consumption_predicted = model.predict(PolynomialFeatures(degree=2).fit_transform([[2]]))

print(f"energy consumption predicted {energy_consumption_predicted}")
```

```python
predicted_existing = model.predict(PolynomialFeatures(degree=2).fit_transform(x))

sns.scatterplot(data=df, x='Temperature',y='Energy Consumption')
plt.plot(x,predicted_existing, color='red')
```
